### 1. 􏰀データ1~4の相関係数を計算し, その結果からわかることを論ぜよ

```python
#データの生成

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n= 1000
x = np.random.randn(1, n)
y1 =  x + np.random.randn(1, n)
y2 =  -x + np.random.randn(1, n)
y3 = np.random.randn(1, n)
y4 = np.power(x, 2) + np.random.randn(1, n)

np.shape(x)
```

```python
#相関係数の表示
print("cor(x, y1)は")
print(np.corrcoef(x, y1)[0,1])
print("cor(x, y2)は")
print(np.corrcoef(x, y2)[0,1])
print("cor(x, y3)は")
print(np.corrcoef(x, y3)[0,1])
print("cor(x, y4)は")
print(np.corrcoef(x, y4)[0,1])
```

### 考察
上記の相関係数（＋宿題上のplot）から判ることは相関係数は比例の関係を表すことは出来るがデータが互いに独立な場合や比例では表せない関係を表すことは出来ない。

### 2. 正規分布の分散共分散行列を変化させると,
確率密度関数の形がどうに変化するか論ぜよ.(解き終わらず)

```python
#平均
mu1 = [-2,-2]
#共分散
cov = [[2,0.5],[0.5,4]]
 
#500はデータ数
x,y = np.random.multivariate_normal(mu1,cov,500).T
```

```python
plt.scatter(x, y)
plt.show()
```

```python
#固有値分解
la, v = np.linalg.eig(cov)
print(la)
v0= v[0][:, np.newaxis]
v1 = v[1][:, np.newaxis]
print(v0)
print(v1)
```

```python
la @ v@np.linalg.inv(v)
```

```python
la[0]*v0@v0.T + la[1]*v1@v1.T
```

```python
#平均
mu1 = [-2,-2]
#共分散
cov = [[2,1],[1,4]]
 
#500はデータ数
x,y = np.random.multivariate_normal(mu1,cov,500).T
plt.scatter(x, y)
plt.show()
```
