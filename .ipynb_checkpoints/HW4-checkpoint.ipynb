# HW4-2 ガウス混合モデルのEMアルゴリズムを実装し、適当な一次元の確率密度関数を推定せよ


必要なもののインポート

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
!python3 --version
%matplotlib inline
```

サンプルデータの生成とプロット

```python
# make data
mu = [13.0, -5.0]
sd = [7.0, 6.0]
N = 10000
data = np.random.normal(mu[0], sd[0], N)
data2 = np.random.normal(mu[1], sd[1], N)
data3 = np.r_[data,data2]
print("データの長さ: " +str(len(data3)))
```

```python
#plot data
plt.hist(data, bins=50, normed=True, alpha=0.5)
plt.hist(data2, bins=50, normed=True, alpha=0.5, color ="red")
plt.xlabel("x")
plt.ylabel("Probability")
plt.show()

plt.hist(data3, bins=50, normed=True, alpha=0.5)
plt.xlabel("x")
plt.ylabel("Probability")
plt.show()
```

初期値の設定

```python
#define initial vals
mu = [random.choice(data3), random.choice(data3)]
temp_sd  = math.sqrt(np.var(data3))
sd = [temp_sd , temp_sd ]
w=[0.5, 0.5]
t = 100 #EMアルゴリズムの反復回数

# lh = [] #尤度の保存用
```

EMアルゴリズムの実行（収束判定はせずに十分な回数アルゴリズムを実行することで代用）

```python
#平均m, 分散vのガウス分布
def gaussian(x, mu, sd):
    p = math.exp(- pow(x - mu, 2) / (2 *pow(sd,2))) / math.sqrt(2 * math.pi * pow(sd,2))
    return p

#E step
def e_step(data, mu, sd,w,  i): #wはスカラー
    eta = [] #負担率の配列
    for x in data:
        d = w[0] * gaussian(x, mu[0], sd[0]) +(1-w[0]) * gaussian(x, mu[1], sd[1])
        n = w[i] * gaussian(x, mu[i], sd[i])
        eta.append(n/d)
    return eta

#M step　関数化していません
def m_step(data, eta):
    return


for i in range(t):
    #Eステップ
    etas = [] #負担率を格納する配列
    for i, k in enumerate(w):
        eta = e_step(data3, mu, sd, w , i)
        etas.append(eta)

        
    #Mステップ    

    new_ws =[]
    new_mus =[]
    new_sds =[]
    #update eta
    for eta in etas:
        new_w  = sum(eta)/len(data3)
        new_ws.append(new_w)
        
    #update mu
        eta = np.array(eta)
        new_mu = (eta.T@data3)/sum(eta)
        new_mus.append(new_mu)
        
    #update sd     
        new_sd =(eta.T @ (data3-new_mu )**2)/(sum(eta)*1)
        new_sd = math.sqrt(new_sd)
        new_sds.append(new_sd)
    w=new_ws
    mu = new_mus
    sd = new_sds
    
#最終結果を表示
print("推定されたパラメータ")
print("重み ", w)  
print("平均", mu)
print("sd ", sd)

```

推定値を用いた分布の確率密度関数の描画

```python
#plot data

#混合分布の用意
xs = np.linspace(min(data3), max(data3), 2*N)
norm1 = mlab.normpdf(xs, mu[0], sd[0])
norm2 = mlab.normpdf(xs, mu[1], sd[1])

plt.hist(data, bins=50, normed=True, alpha=0.5)
plt.hist(data2, bins=50, normed=True, alpha=0.5, color ="red")
plt.plot(xs, w[0] * norm1 + w[1] * norm2, color="red", lw=3)#混合分布のプロット
plt.xlabel("x")
plt.ylabel("Probability")
plt.show()

plt.hist(data3, bins=50, normed=True, alpha=0.5)
plt.plot(xs, w[0] * norm1 + w[1] * norm2, color="red", lw=3)#混合分布のプロット
plt.xlabel("x")
plt.ylabel("Probability")
plt.show()
```
